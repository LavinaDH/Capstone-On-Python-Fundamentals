```python
# Task 1: Creating 03 dataframes and saving in CSV for further tasks

import pandas as pd
import numpy as np

#------------
# Employee dataframe as pre PDF
#------------
 
employee_data = {
    "ID":["A001","A002","A003","A004","A005"],
    "Name": ["John Alter", "Alice Luxumberg", "Tom Sabestine", "Nina Adgra", "Amy Johny"],
    "Gender": ["M", "F", "M", "F", "F"],
    "City": ["Paris", "London", "Berlin", "Newyork", "Madrid"],
    "Age": [25,27,29,31,30]
}
df_employee = pd.DataFrame(employee_data)
# Save Employee to CSV
df_employee.to_csv("Employee.csv", index=False)

#----------
# Seniority Level DataFrame
#----------

seniority_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005"],
    "Designation_Level": [ 2,2,3,2,3]
}
df_seniority= pd.DataFrame(seniority_data)
#Saving Seniority in CSV
df_seniority.to_csv("Seniourity.csv", index=False)

#-----------
#Project DataFrame
#-----------

project_data = {
    "ID": ["A001","A002","A003","A004","A005",
           "A002","A005","A003","A001","A003",
           "A001","A004","A004","A005"],
    "Project": ["Project 1", "Project 2", "Project 3", "Project 4",  "Project 5",
                "Project 6","Project 7", "Project 8", "Project 9", "Project 10",
                "Project 11", "Project 12", "Project 13", "Project 14"],
    "Cost": [1002000,2000000,4500000,5500000, np.nan,
             680000,400000,350000, np.nan,300000,
             2000000,1000000,3000000,200000],
    "Status": ["Finished", "Ongoing", "Finished", "Ongoing", "Finished",
               "Failed", "Finished", "Failed", "Ongoing", "Finished",
               "Failed", "Ongoing", "Finished", "Finished"]
}
df_project = pd.DataFrame(project_data)
#Saving Project in CSV
df_project.to_csv("Project.csv", index=False)

print("Employee DataFrame:")
display(df_employee)

print("\nseniority DataFrame:")
display(df_seniority)

print("\nproject DataFrame:")
display(df_project)
```

```python
#Task 2: Replacing missing values in the cost column of the dataframe "Project" with running average using the for loop.

import pandas as pd
import numpy as np

proj = pd.read_csv("Project.csv")

#Creating a List to keep a trace of non-missing cost values.

seen_costs = []

#Loop through each row in the table.

for i in range (len(proj)):
    cost = proj.loc[i,"Cost"]

    #if Cost is present
    if pd.notna(cost):
        seen_costs.append(cost)

    #if cost is missing but has a previous value, apply running avg and round it to 0
    elif seen_costs:
        running_avg = sum(seen_costs)/len(seen_costs)
        proj.loc[i,"Cost"] = round(running_avg,0)

    #if cost is missing and this is the first row (no previous values)
    else:
        proj.loc[i,"Cost"] = 0

 #Saving this as a Fresh CSV, to keep the original data

display(proj)
proj.to_csv("Project_filled.csv", index=False)
print(" Project_filled.csv saved successfully!")
```

```python
# Task 3: Split "First Name" and "Second Name" from the "Name" Column in the Employee Dataframe and Drop the "name Coloum

import pandas as pd

emp = pd.read_csv("Employee.csv")

# Split 'Name' into 'First Name' and 'Last Name'
emp["First Name"] = emp["Name"].apply(lambda x: x.split()[0])                # before the space
emp["Last Name"] = emp["Name"].apply(lambda x: " ".join(x.split()[1:]))     # after the space

# Drop the old 'Name' column
emp = emp.drop(columns=["Name"])

# Display the updated DataFrame
display(emp)

# Save it as a new file so we keep the original safe
emp.to_csv("Employee_split.csv", index=False)
print("Task 3 complete — Employee_split.csv saved successfully!")
```

```python
# TASK 4: Join all three DataFrames into a single Final DataFrame
# Joining them all in using the common column "ID"

import pandas as pd

# Read the 3 CSV files.

emp = pd.read_csv("Employee_split.csv")     # from Task 3
sen = pd.read_csv("Seniourity.csv")          # from Task 1
proj = pd.read_csv("Project_filled.csv")    # from Task 2

# Merging Employee and Seniority on 'ID' first
emp_sen = emp.merge(sen, on="ID", how="left")

# Now merging the above with the Project on 'ID'
Final = proj.merge(emp_sen, on="ID", how="left")

# Reorder the columns to make it look neat
cols_order = ["ID", "First Name", "Last Name", "Gender", "City", "Age",
              "Designation_Level", "Project", "Cost", "Status"]
Final = Final[cols_order]

# Display the final merged DataFrame
display(Final)

# Save the final combined table
Final.to_csv("Final.csv", index=False)
print("Task 4 complete — Final.csv saved successfully!")
```

```python
#Task 5: Adding a 5% bonus to Employees who have finished their projects.

import pandas as pd

# Read the Final.csv file created in Task 4
Final = pd.read_csv("Final.csv")

# Calculate bonus: 5% of Cost if Status == 'Finished', else 0
Final["bonus"] = Final.apply(
    lambda row: row["Cost"] * 0.05 if str(row["Status"]).strip().lower() == "finished" else 0,
    axis=1
)

# Round the bonus to the nearest integer
Final["bonus"] = Final["bonus"].round(0)

# Display the updated DataFrame
display(Final)

# Save the updated file
Final.to_csv("Final_with_bonus.csv", index=False)
print("Task 5 complete — Final_with_bonus.csv saved successfully!")
```

```python
# TASK 6: Demote employees by 1 designation level who failed their project, and remove those with a designation of more than 4

import pandas as pd

# Read the latest data from Task 5
Final = pd.read_csv("Final_with_bonus.csv")
sen = pd.read_csv("Seniourity.csv")

# STEP 1️ — Count how many failed projects each employee has
fail_counts = Final[Final["Status"].str.lower() == "failed"].groupby("ID").size().to_dict()
print("Failed project count per employee:", fail_counts)

# STEP 2️ — Set 'ID' as the index to update the seniority table easily
sen = sen.set_index("ID")

# STEP 3️ — For each failed employee, demote by 1 designation level
for emp_id, count in fail_counts.items():
    if emp_id in sen.index:
        sen.loc[emp_id, "Designation_Level"] = sen.loc[emp_id, "Designation_Level"] + count       
       
# STEP 4️ — Remove employees whose designation level is now > 4
sen_after = sen[sen["Designation_Level"] <= 4].reset_index()

# STEP 5️ — Save the updated seniority list
sen_after.to_csv("Seniority_after_demote.csv", index=False)
print("\n Updated Seniority Table saved as Seniority_after_demote.csv")
display(sen_after)

# STEP 6️ — Update the Final table too (keep only remaining employees)
Final_after = Final.merge(sen_after, on="ID", how="inner", suffixes=("", "_updated"))

# Replace the old Designation_Level with the updated one
Final_after = Final_after.drop(columns=["Designation_Level"]).rename(columns={"Designation_Level_updated": "Designation_Level"})

# Save the updated Final table
Final_after.to_csv("Final_after_task6.csv", index=False)
print("\n Final table updated and saved as Final_after_task6.csv")
display(Final_after)
```

```python
# TASK 7: Add Mr. or Mrs. to First Name based on Gender, then drop the Gender column

import pandas as pd

# Read the latest file
Final = pd.read_csv("Final_after_task6.csv")

# Add prefix based on Gender
Final["First Name"] = Final.apply(
    lambda row: ("Mr. " + row["First Name"]) if str(row["Gender"]).strip().lower() == "m"
    else ("Mrs. " + row["First Name"]),
    axis=1
)

# Drop the Gender column
Final = Final.drop(columns=["Gender"])

# Display the updated table
display(Final)

# Save the updated version
Final.to_csv("Final_with_titles.csv", index=False)
print("Task 7 complete — Final_with_titles.csv saved successfully!")
```

```python
# TASK 8: Promote designation level by 1 if Age > 29 using IF condition

import pandas as pd

# Read the latest dataset
Final = pd.read_csv("Final_with_titles.csv")

# Apply promotion rule using IF condition
Final["Designation_Level"] = Final.apply(
    lambda row: row["Designation_Level"] + 1 if row["Age"] > 29 else row["Designation_Level"],
    axis=1
)

# Display the updated table
display(Final)

# Save the new file
Final.to_csv("Final_after_task8.csv", index=False)
print(" Task 8 complete — Final_after_task8.csv saved successfully!")
```

```python
# TASK 9: Sum the total project cost for each employee

import pandas as pd

# Read the latest dataset
Final = pd.read_csv("Final_after_task8.csv")

# Group by ID and First Name, then sum up the Cost
TotalProjCost = Final.groupby(["ID", "First Name"])["Cost"].sum().reset_index()

# Rename the column
TotalProjCost.rename(columns={"Cost": "Total Cost"}, inplace=True)

# Display the new DataFrame
display(TotalProjCost)

# Save it to a new file
TotalProjCost.to_csv("TotalProjCost.csv", index=False)
print(" Task 9 complete — TotalProjCost.csv saved successfully!")
```

```python
# TASK 10: Print employees whose city name contains the letter 'o'

import pandas as pd

# Read the latest dataset
Final = pd.read_csv("Final_after_task8.csv")

# Filter employees where city contains 'o' (case-insensitive)
city_with_o = Final[Final["City"].str.contains("o", case=False, na=False)]

# Display the filtered result
display(city_with_o)

# Optional: Save this filtered list
city_with_o.to_csv("Employees_City_with_O.csv", index=False)
print(" Task 10 complete — Employees_City_with_O.csv saved successfully!")
```
